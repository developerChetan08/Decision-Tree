```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
iris.data
```

```python
iris.target
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
X = df.iloc[:,:-1]
y = iris.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(model,filled=True)
plt.savefig("Tree.png")
```

# Postpruning:- [max-depth]

```python
post_model = DecisionTreeClassifier(max_depth=2)
```

```python
post_model.fit(X_train,y_train)
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(post_model,filled=True)
```

```python
##Predicition
y_pred = post_model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score = accuracy_score(y_pred,y_test)
score
```

```python
print(classification_report(y_pred,y_test))
```

# Preprunning[Mixed All Parameter]:-

```python
parameter={
 'criterion':['gini','entropy','log_loss'],
  'splitter':['best','random'],
  'max_depth':[1,2,3,4,5],
  'max_features':['auto', 'sqrt', 'log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
mixed_model = DecisionTreeClassifier()
cv = GridSearchCV(mixed_model,param_grid=parameter,cv=5,scoring='accuracy')
```

```python
cv.fit(X_train,y_train)
```

```python
cv.best_params_
```

```python
y_pred = cv.predict(X_test)
y_pred
```

```python
score = accuracy_score(y_pred,y_test)
score
```

```python
print(classification_report(y_pred,y_test))
```
